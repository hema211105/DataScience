```python
for i in df.select_dtypes(include=['int','float']).columns:
  df.fillna(df[i].median(),inplace=True)
for i in df.select_dtypes(include='object').columns:
  df.fillna(df[i].value_counts().index[0],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

for i in df.select_dtypes(include='object').columns:
    df[i] = df[i].astype(str)
    le=LabelEncoder()
    le.fit(df[i])
    transformed_values=le.transform(df[i])
    df[i]=transformed_values
```

```python
##model building
df=df.sample(10000)
y=df['TARGET']
x=df.drop('TARGET',axis=1)
```

```python
y.shape,x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
print(dt.predict(x_test)[0:10])
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
print(rf.predict(x_test)[0:10])
```

```python
# model validation
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_test,dt.predict(x_test)))
print(confusion_matrix(y_test,rf.predict(x_test)))
print(accuracy_score(y_test,dt.predict(x_test)))
print(accuracy_score(y_test,rf.predict(x_test)))
```
