```python
from sklearn.metrics import mean_absolute_error
def model_prediction(algoname,x_train,y_train,x_test,y_test):
    model=algoname
    model.fit(x_train,y_train)
    predicted_values=model.predict(x_test)
    actual_values=y_test.values
    mae=sum(abs(actual_values-predicted_values))/(x_test.shape[0])
    mae=mean_absolute_error(actual_values,predicted_values)
    mse=np.square(mae)
    rmse=np.sqrt(mse)
    mape=((actual_values-predicted_values)/actual_values).sum()/x_test.shape[0]
    return [mae,mse,rmse,mape]
```

```python
model_prediction(DecisionTreeRegressor(),x_train,y_train,x_test,y_test)
```

```python
model_prediction(RandomForestRegressor(),x_train,y_train,x_test,y_test)
```

```python
model_prediction(LinearRegression(),x_train,y_train,x_test,y_test)
```
